```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=10)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
reg_pred = regression.predict(X_test)
```

```python
plt.scatter(y_test,reg_pred)
plt.xlabel("test truth data")
plt.ylabel("Test Predicted data")
```

```python
#Assumptions - 
#1 Linear relationship between predicted value and truth value
#2 Distribution plot of residuals should be normal distribution
#3 Distribution between predictions and residuals should be uniform distribution. known as honmosedacity
```

```python
residuals = y_test - reg_pred
```

```python
sns.displot(residuals,kind="kde")
```

```python
#Scatter plot with predictions and residulas
#uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
#Performance Metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
##R-Square
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

```python
1 - ( 1 - score)* (len(y_test) - 1)/(len(y_test) - X_test.shape[1] - 1 )
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train,y_train)
print(ridge.coef_)
print(ridge.intercept_)
ridge_pred = ridge.predict(X_test)
plt.scatter(y_test,ridge_pred)
plt.xlabel("test truth data")
plt.ylabel("Test Predicted data")
residuals1 = y_test - reg_pred
sns.displot(residuals1,kind="kde")
#uniform distribution
plt.scatter(reg_pred,residuals1)
#Performance Metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test,ridge_pred))
print(mean_absolute_error(y_test,ridge_pred))
print(np.sqrt(mean_squared_error(y_test,ridge_pred)))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train,y_train)
print(lasso.coef_)
print(lasso.intercept_)
lasso_pred = lasso.predict(X_test)
plt.scatter(y_test,lasso_pred)
plt.xlabel("test truth data")
plt.ylabel("Test Predicted data")
residuals2 = y_test - lasso_pred
sns.displot(residuals2,kind="kde")
#uniform distribution
plt.scatter(lasso_pred,residuals2)
#Performance Metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test,lasso_pred))
print(mean_absolute_error(y_test,lasso_pred))
print(np.sqrt(mean_squared_error(y_test,lasso_pred)))
```
